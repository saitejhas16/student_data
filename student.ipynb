## DATA ANALYSIS ON STUDENT DATA

```python
##importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
##Loading student dataset
df=pd.read_csv("student_data.csv")
df
```

```python
##changing student id into range of numbers
df['Student ID']=[i for i in range(1,40001)]
df
```

```python
##changing column name of "passed" to "result"
df=df.rename(columns={'Passed':'Result'})

## replacing values of result column of yes and no with pass and fail respectively
df.loc[df['Result']=='Yes','Result']='Pass'
df.loc[df['Result']=='No','Result']='Fail'

df
```

```python
## replacing "NaN" values in 3 columns with 0
df[['Study Hours per Week', 'Attendance Rate', 'Previous Grades']] = df[['Study Hours per Week', 'Attendance Rate', 'Previous Grades']].fillna(0)
df
```

```python
# Rounding off the values of the 3 columns and converting them to integer data type
df[['Study Hours per Week', 'Attendance Rate', 'Previous Grades']] = df[['Study Hours per Week', 'Attendance Rate', 'Previous Grades']].round().astype(int)
df
```

```python
##saving cleaned and modified data
df.to_csv("student_modified.csv",
```

## Visualization of Student Attendance Rates Distribution

```python
# Define the bins and labels
bins = [-float('inf'), 65, 75, float('inf')]
labels = ['Below 65', '65 to 75', 'Above 75']

# Categorize the attendance rates into bins
df['category'] = pd.cut(df['Attendance Rate'], bins=bins, labels=labels)

# Count the number of students in each category
counts = df['category'].value_counts()

print(counts)
```

```python
counts.plot(kind="pie",autopct='%1.1f%%',wedgeprops=dict(width=0.5))
plt.title('Distribution of Attendance Rates',fontsize=18)
plt.show()
```

## Visualisation of Grades among Students

```python
# Defining the bins and labels
bins = [0, 40, 50, 60, 75, 85, 90, 100,200]
labels = ['F', 'E', 'D', 'C', 'B', 'B+', 'A','A+']

# Categorizing the grades into bins
df['category1'] = pd.cut(df['Previous Grades'], bins=bins, labels=labels)

# Counting the number of students in each category
counts1 = df['category1'].value_counts()

print(counts1)
```

```python
plt.figure(figsize=(7,7))
counts1.plot(kind="pie",autopct='%1d%%')
plt.title('Distribution of Grades among students',fontsize=18)
plt.show()
```

## Visualization of Student Counts in Extracurricular Activities

```python
df_pa=df['Participation in Extracurricular Activities'].value_counts()
df_pa
```

```python
# Create a horizontal bar chart
plt.figure(figsize=(8, 6))
df_pa.plot(kind='barh', color=['blue', 'green'])
plt.xlabel('Number of Students')  
plt.ylabel('Participation Status') 
# Set x-axis limits and ticks
plt.xlim(18000, df_pa.max() + 250)
plt.xticks(range(18000, df_pa.max() + 1000, 250))  #Ticks to start from 1700 and increase by 250

# Set axis labels and title
plt.title('Count of Participation in Extracurricular Activities')

# Show the plot
plt.show()
```

## Visualisation of Parents education level

```python
df_p=df.groupby('Parent Education Level')['Result'].count()
df_p
```

```python
df_p.plot(kind='bar')

# Getting the maximum value in the Series
max_value = df_p.max()
plt.title("Education levels of parents v/s count ",fontsize=15)
# Setting y-axis ticks starting from 7000 with a difference of 50
plt.yticks(range(7000, int(max_value) + 50, 50))
plt.ylabel("count of parents",fontsize=12)
# Setting the y-axis limit to start from 7000
plt.ylim(bottom=7000)
plt.show()
```
